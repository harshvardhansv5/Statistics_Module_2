**1. What is hypothesis testing in statistics?**
- Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample to support a particular claim about a population parameter. It involves comparing observed data to expected results under a null hypothesis.

 Example: Testing whether a new teaching method improves exam scores by comparing student performance to a known average.

**2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**
- Null Hypothesis (H₀): Assumes no effect or difference (e.g., H₀: μ = 100).

- Alternative Hypothesis (H₁): Suggests an effect or difference (e.g., H₁: μ ≠ 100).

 The alternative hypothesis is the claim being tested, while the null hypothesis represents the status quo.

**3. What is the significance level in hypothesis testing, and why is it important?**
- Significance Level (α) - The probability of rejecting H₀ when it is actually true (Type I error). A common choice is α = 0.05, indicating a 5% risk of mistakenly rejecting H₀.

 Example: If α = 0.05, results yielding a p-value below 0.05 are considered statistically significant.

**4. What does a P-value represent in hypothesis testing?**
- The probability of observing the sample data, or more extreme results, assuming H₀ is true. A lower p-value suggests stronger evidence against H₀.

 Example: P = 0.01 means there is only a 1% chance of obtaining the observed data under H₀.

**5. How do you interpret the P-value in hypothesis testing?**
- If $P ≤ α$, reject H₀ (support for H₁).

- If $P > α$, fail to reject H₀ (insufficient evidence to support H₁).

 Example: P = 0.03 at α = 0.05 → Reject H₀ (statistically significant result).

**6. What are Type 1 and Type 2 errors in hypothesis testing?**
- Type I Error (α): Rejecting a true H₀ (false positive).

- Type II Error (β): Failing to reject a false H₀ (false negative).

 Example:

 Type I: Diagnosing a healthy person with a disease.

 Type II: Failing to detect a disease in a sick person.

**7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**
- One-tailed: Tests for an effect in one direction (e.g., H₁: μ > 50).

- Two-tailed: Tests for any significant difference (e.g., H₁: μ ≠ 50).

 Example: Testing if a drug increases recovery time (one-tailed) vs. testing if it affects recovery time in any way (two-tailed).

**8. What is the Z-test, and when is it used in hypothesis testing?**
- Used when the population variance is known or when sample size is large (n ≥ 30). It follows a standard normal distribution.

 Example: Testing if a sample mean (n=50) differs from a known population mean when σ is known.

**9. How do you calculate the Z-score, and what does it represent in hypothesis testing?**
- Formula: $Z = (X̄ − μ) / (σ/√n)$                                      
  Represents the number of standard errors X̄ is away from μ.

 Example: Z = 2.5 implies the sample mean is 2.5 SEs above the population mean

**10. What is the T-distribution, and when should it be used instead of the normal distribution?**
- Used when the population standard deviation (σ) is unknown and the sample size is small (n < 30). It accounts for added uncertainty.

 Example: Testing a sample mean with n = 20 when σ is unknown.

**11. What is the difference between a Z-test and a T-test?**
- Z-test: Used when σ is known and/or n ≥ 30.

- T-test: Used when σ is unknown and n < 30 (uses sample standard deviations).

**12. What is the T-test, and how is it used in hypothesis testing?**
- A statistical test used to compare means and determine if differences are significant. It is used when the population standard deviation (σ) is unknown.

 Types:-

- One-sample T-test: Compares a sample mean to a known population mean.

- Independent (Unpaired) T-test: Compares means of two unrelated groups. Assumes equal or unequal variances.

- Paired T-test: Compares means of the same group before and after a treatment.

 Example: Evaluating weight loss before and after a diet program (paired t-test).

**13. What is the relationship between Z-test and T-test in hypothesis testing?**
- The **Z-test and T-test** are both statistical methods used in **hypothesis testing** to compare sample means with population parameters or between groups. Their primary relationship is that they both test for differences in means but are used under different conditions:

### **Key Differences and Relationship**  

1. **Sample Size (n):**
   - **Z-test** is used when the sample size is **large (n ≥ 30)**.
   - **T-test** is used when the sample size is **small (n < 30)**.

2. **Population Standard Deviation (σ):**
   - **Z-test** requires the population **standard deviation (σ) to be known**.
   - **T-test** is used when the population **standard deviation is unknown** and must be estimated from the sample.

3. **Distribution Used:**
   - **Z-test** follows the **standard normal distribution (Z-distribution)**.
   - **T-test** follows the **t-distribution**, which has **heavier tails** (accounts for more variability in small samples).

4. **Convergence of T-test to Z-test:**
   - As the sample size **increases (n → ∞)**, the **t-distribution approaches the normal (Z) distribution**.
   - This means that for **large samples**, a **T-test can approximate a Z-test**, making them interchangeable.

**20. What is an ANOVA test, and what are its assumptions?**
- ### **ANOVA (Analysis of Variance) Test**  
ANOVA is a **statistical method** used to compare the means of three or more groups to determine if there is a significant difference between them. It prevents the inflation of Type I errors that would occur if multiple **t-tests** were performed.  

#### **Types of ANOVA:**  
1. **One-way ANOVA:** Compares means of one independent variable across multiple groups.  
   *Example:* Testing if three different fertilizers affect plant growth differently.  
   
2. **Two-way ANOVA:** Analyzes the effect of two independent variables and their interaction.  
   *Example:* Studying the effect of **diet type** and **exercise level** on weight loss.  
   
3. **Repeated Measures ANOVA:** Tests the same subjects under different conditions over time.  
   *Example:* Measuring student performance before, during, and after a training program.  



### **Assumptions of ANOVA**  
For ANOVA to be valid, the following assumptions must hold:  

1. **Independence:**  
   - Observations must be independent within and across groups.  
   - *Example:* Each student’s exam score should not be influenced by another student’s score.  

2. **Normality:**  
   - The data in each group should be approximately normally distributed.  
   - *Check:* Use **Shapiro-Wilk test** or **Q-Q plots** to assess normality.  

3. **Homogeneity of Variances (Homoscedasticity):**  
   - The variance across groups should be equal.  
   - *Check:* Use **Levene’s test** or **Bartlett’s test** to verify.  

4. **No significant outliers:**  
   - Extreme values can distort results.  
   - *Check:* Use **box plots** or **Z-scores** to detect outliers.  

If these assumptions are **violated**, alternatives like the **Kruskal-Wallis test (non-parametric ANOVA)** can be used.

**21. What are the different types of ANOVA tests?**
- Types of ANOVA
 - One-way: Tests one independent variable.
 - Two-way: Tests two independent variables and interactions.
 - Repeated Measures: Tests the same subjects under multiple conditions.

**22. What is the F-test, and how does it relate to hypothesis testing?**
- Uses the F-statistic (variance ratio) to compare groups. Applied in ANOVA and regression.

 Example: Determining whether regression model coefficients are jointly significant.

**1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results?**

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Compute the Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Compute the p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

    # Interpretation
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between the sample and population mean.")

# Example usage
sample_data = [52, 48, 50, 47, 49, 53, 51, 46, 50, 52]
population_mean = 50
population_std = 3  # Known population standard deviation
alpha = 0.05  # Significance level

z_test(sample_data, population_mean, population_std, alpha)
```

**4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Compute the Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Compute the p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

    # Interpretation
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between the sample and population mean.")

    # Visualization of decision regions
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    critical_value = norm.ppf(1 - alpha/2)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='Standard Normal Distribution')
    plt.fill_between(x, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.3, label='Rejection Region')
    plt.axvline(z_score, color='blue', linestyle='dashed', label=f'Z-score: {z_score:.2f}')
    plt.axvline(-critical_value, color='black', linestyle='dashed', label=f'Critical Value: ±{critical_value:.2f}')
    plt.axvline(critical_value, color='black', linestyle='dashed')

    plt.legend()
    plt.title('Two-Tailed Z-Test Decision Region')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.show()

# Simulating random sample data
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=50, scale=3, size=30)  # Sample from normal distribution

population_mean = 50
population_std = 3  # Known population standard deviation
alpha = 0.05  # Significance level

z_test(sample_data, population_mean, population_std, alpha)
```

**5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Compute the Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Compute the p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

    # Interpretation
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between the sample and population mean.")

    # Visualization of decision regions
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    critical_value = norm.ppf(1 - alpha/2)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='Standard Normal Distribution')
    plt.fill_between(x, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.3, label='Rejection Region')
    plt.axvline(z_score, color='blue', linestyle='dashed', label=f'Z-score: {z_score:.2f}')
    plt.axvline(-critical_value, color='black', linestyle='dashed', label=f'Critical Value: ±{critical_value:.2f}')
    plt.axvline(critical_value, color='black', linestyle='dashed')

    plt.legend()
    plt.title('Two-Tailed Z-Test Decision Region')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.show()

def visualize_errors(alpha=0.05, beta=0.2):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    critical_value = norm.ppf(1 - alpha/2)
    type_2_shift = 1.5  # Assume alternative hypothesis shifts mean
    y_alt = norm.pdf(x, type_2_shift, 1)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='Null Distribution (H0)', color='blue')
    plt.plot(x, y_alt, label='Alternative Distribution (H1)', color='green')

    plt.fill_between(x, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.3, label='Type 1 Error (Alpha)')
    plt.fill_between(x, y_alt, where=(x > -critical_value) & (x < critical_value), color='purple', alpha=0.3, label='Type 2 Error (Beta)')

    plt.axvline(-critical_value, color='black', linestyle='dashed', label=f'Critical Value: ±{critical_value:.2f}')
    plt.axvline(critical_value, color='black', linestyle='dashed')

    plt.legend()
    plt.title('Type 1 and Type 2 Errors')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.show()

# Simulating random sample data
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=50, scale=3, size=30)  # Sample from normal distribution

population_mean = 50
population_std = 3  # Known population standard deviation
alpha = 0.05  # Significance level

z_test(sample_data, population_mean, population_std, alpha)
visualize_errors(alpha=0.05, beta=0.2)
```

**6. Write a Python program to perform an independent T-test and interpret the results.**

```python
import numpy as np
from scipy import stats

# Set random seed for reproducibility
np.random.seed(42)

# Generate two independent sample datasets
group1 = np.random.normal(loc=5, scale=1.5, size=30)  # Mean = 5, SD = 1.5
group2 = np.random.normal(loc=6, scale=1.5, size=30)  # Mean = 6, SD = 1.5

# Check homogeneity of variances using Levene's test
levene_stat, levene_p = stats.levene(group1, group2)
alpha = 0.05  # Significance level

# Perform appropriate T-test based on variance check
if levene_p > alpha:
    print("Equal variances assumed (p = {:.3f})".format(levene_p))
    t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=True)
else:
    print("Equal variances not assumed (p = {:.3f})".format(levene_p))
    t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=False)

# Output results
print("\nT-test Results:")
print("T-statistic: {:.3f}".format(t_stat))
print("P-value: {:.3f}".format(p_value))

# Interpret the results
print("\nInterpretation:")
if p_value < alpha:
    print("The p-value is less than 0.05. Reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference between the two groups.")
else:
    print("The p-value is greater than or equal to 0.05. Fail to reject the null hypothesis.")
    print("Conclusion: No significant difference between the two groups.")
```

**9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.**

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    sample (array-like): The sample data.
    confidence (float): The confidence level (default = 95%).

    Returns:
    tuple: (Lower bound, Upper bound) of the confidence interval.
    """
    sample_size = len(sample)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (unbiased)

    # Compute standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Compute the margin of error using t-distribution (for small samples)
    t_critical = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)
    margin_of_error = t_critical * standard_error

    # Compute confidence interval bounds
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Simulate sample data
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=5, size=30)  # Sample from normal distribution

# Compute 95% confidence interval
ci_lower, ci_upper = confidence_interval(sample_data)

# Print results
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

**10. Write a Python program to calculate the margin of error for a given confidence level using sample data.**

```python
import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for a given confidence level.

    Parameters:
    sample (array-like): The sample data.
    confidence (float): The confidence level (default = 95%).

    Returns:
    float: The margin of error.
    """
    sample_size = len(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (unbiased)

    # Compute standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Compute the critical value using t-distribution (for small samples)
    t_critical = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)

    # Compute margin of error
    margin_of_error = t_critical * standard_error

    return margin_of_error

# Simulate sample data
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=5, size=30)  # Sample from normal distribution

# Compute margin of error for 95% confidence level
moe = margin_of_error(sample_data)

# Print results
print(f"Margin of Error (95% Confidence Level): {moe:.2f}")
```

**11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.**

```python
def bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A):
    # Total probability of B (Law of Total Probability)
    evidence_B = (likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * prior_not_A)

    # Compute posterior probability
    posterior_A_given_B = (likelihood_B_given_A * prior_A) / evidence_B

    return posterior_A_given_B

# Example Scenario:
# Suppose we are testing for a rare disease:
prior_disease = 0.01  # 1% of people have the disease
likelihood_positive_given_disease = 0.95  # Test correctly identifies 95% of cases
prior_no_disease = 1 - prior_disease  # 99% of people do not have the disease
likelihood_positive_given_no_disease = 0.05  # 5% false positive rate

# Compute probability of having the disease given a positive test result
posterior_probability = bayes_theorem(
    prior_disease,
    likelihood_positive_given_disease,
    prior_no_disease,
    likelihood_positive_given_no_disease
)

print(f"Probability of having the disease given a positive test: {posterior_probability:.4f}")
```

**16. Implement an F-test using Python to compare the variances of two random samples.**

```python
import numpy as np
from scipy.stats import f

# Simulate two random samples
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=5, size=30)  # Mean = 50, SD = 5
sample2 = np.random.normal(loc=50, scale=8, size=30)  # Mean = 50, SD = 8

# Compute sample variances
var1 = np.var(sample1, ddof=1)  # Sample variance (unbiased)
var2 = np.var(sample2, ddof=1)

# Ensure F-statistic is always ≥ 1
if var1 < var2:
    var1, var2 = var2, var1  # Swap variances

# Calculate F-statistic
F_statistic = var1 / var2

# Degrees of freedom
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# Compute p-value (two-tailed test)
p_value = 2 * min(f.cdf(F_statistic, df1, df2), 1 - f.cdf(F_statistic, df1, df2))

# Print results
print(f"Variance of Sample 1: {var1:.3f}")
print(f"Variance of Sample 2: {var2:.3f}")
print(f"F-statistic: {F_statistic:.3f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

**19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def check_anova_assumptions(*groups):
    """Checks ANOVA assumptions: normality, independence, and homogeneity of variances."""

    # Normality test (Shapiro-Wilk test)
    for i, group in enumerate(groups, start=1):
        stat, p = stats.shapiro(group)
        print(f"Shapiro-Wilk Test for Group {i}: W-statistic={stat:.4f}, p-value={p:.4f}")
        if p < 0.05:
            print(f"Group {i} may not be normally distributed.")
        else:
            print(f"Group {i} appears normally distributed.")

    # Homogeneity of variances test (Levene's test)
    stat, p = stats.levene(*groups)
    print(f"\nLevene's Test for Equal Variances: W-statistic={stat:.4f}, p-value={p:.4f}")
    if p < 0.05:
        print("Variances are not equal across groups.")
    else:
        print("Variances are equal across groups.")

    print("\nIndependence is usually checked based on study design.")

# Simulating random data for three groups
group_1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10, Size=30
group_2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std=10, Size=30
group_3 = np.random.normal(loc=60, scale=10, size=30)  # Mean=60, Std=10, Size=30

# Checking ANOVA assumptions
check_anova_assumptions(group_1, group_2, group_3)

# Performing One-Way ANOVA test
f_stat, p_value = stats.f_oneway(group_1, group_2, group_3)

# Interpreting the results
alpha = 0.05
print(f"\nF-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < alpha:
    print("Reject the null hypothesis: At least one group mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between group means.")

# Visualizing the results with boxplots
plt.boxplot([group_1, group_2, group_3], labels=['Group 1', 'Group 2', 'Group 3'])
plt.ylabel('Values')
plt.title('One-Way ANOVA Test - Group Comparisons')
plt.show()
```

**24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.**

```python
import numpy as np
import scipy.stats as stats

# Simulating data: Random sample from a normal distribution
np.random.seed(42)  # For reproducibility
sample = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10, n=30

# Hypothesized population variance (sigma_0^2)
sigma_0_sq = 100  # Assume population variance is 100

# Sample variance
sample_variance = np.var(sample, ddof=1)  # Unbiased sample variance (n-1)

# Chi-square test statistic
n = len(sample)
chi2_stat = (n - 1) * sample_variance / sigma_0_sq

# Compute p-value (two-tailed test)
p_value = 2 * min(stats.chi2.cdf(chi2_stat, df=n-1), 1 - stats.chi2.cdf(chi2_stat, df=n-1))

# Display results
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Population variance is significantly different from assumed variance.")
else:
    print("Fail to reject the null hypothesis: No significant difference in population variance.")
```

**25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups**.

```python
import numpy as np
import scipy.stats as stats

def z_test_proportions(success1, size1, success2, size2):
    """Performs a two-sample Z-test for proportions."""
    # Proportion estimates
    p1 = success1 / size1
    p2 = success2 / size2

    # Pooled proportion
    p_pool = (success1 + success2) / (size1 + size2)

    # Standard error
    se = np.sqrt(p_pool * (1 - p_pool) * (1/size1 + 1/size2))

    # Z-score
    z_stat = (p1 - p2) / se

    # P-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    return z_stat, p_value

# Example data: Group 1 and Group 2 success counts and sample sizes
success1, size1 = 45, 200  # Group 1: 45 successes out of 200
success2, size2 = 30, 180  # Group 2: 30 successes out of 180

# Perform Z-test
z_stat, p_value = z_test_proportions(success1, size1, success2, size2)

# Display results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in proportions.")
```

**26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def f_test_variance(sample1, sample2):
    """Performs an F-test to compare variances of two datasets."""
    var1 = np.var(sample1, ddof=1)  # Unbiased variance of sample 1
    var2 = np.var(sample2, ddof=1)  # Unbiased variance of sample 2

    # Compute F-statistic
    f_stat = var1 / var2 if var1 > var2 else var2 / var1

    # Degrees of freedom
    dfn = len(sample1) - 1
    dfd = len(sample2) - 1

    # Compute p-value (one-tailed test)
    p_value = 1 - stats.f.cdf(f_stat, dfn, dfd)

    return f_stat, p_value

# Simulating data: Two independent samples from normal distributions
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=50, scale=15, size=30)

# Perform F-test
f_stat, p_value = f_test_variance(sample1, sample2)

# Display results
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")

# Visualization
plt.figure(figsize=(8, 5))
plt.boxplot([sample1, sample2], labels=['Sample 1', 'Sample 2'], patch_artist=True)
plt.title("Comparison of Variances")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```
